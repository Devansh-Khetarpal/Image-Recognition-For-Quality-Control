```python
from ultralytics import YOLO
import io
import numpy as np
import PIL.Image
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
model = YOLO(r'D:\DEVANSH COLLEGE\Btech 7th Sem\New folder\Defect Detection.v6i.yolov8\runs\detect\train\weights\best.pt')
upload_button = widgets.FileUpload(
    accept='image/*',  
    multiple=False      
)
display(upload_button)
def process_and_display_image(uploaded_image):
    image_data = uploaded_image['content']
    image = PIL.Image.open(io.BytesIO(image_data))
    temp_path = 'temp_image.jpg'
    image.save(temp_path)
    results = model.predict(temp_path)
    for result in results:
        result.show()
def on_upload_change(change):
    uploaded_files = upload_button.value
    if isinstance(uploaded_files, dict): 
        for filename, file in uploaded_files.items():
            process_and_display_image(file)
    elif isinstance(uploaded_files, tuple):  
        for file in uploaded_files:
            process_and_display_image(file)
upload_button.observe(on_upload_change, names='value')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('D:\\DEVANSH COLLEGE\\BTech 6th Sem\\Internship\\New folder\\Fisheye Dataset Final\\runs\\detect\\train9\\results.csv')
df.columns = df.columns.str.strip()  
sns.set(style="whitegrid")
fig, axes = plt.subplots(3, 2, figsize=(14, 16))
fig.suptitle('Training Metrics Over Epochs', fontsize=20)
axes[0, 0].plot(df['epoch'], df['metrics/mAP50(B)'], label='mAP50', color='blue')
axes[0, 0].plot(df['epoch'], df['metrics/mAP50-95(B)'], label='mAP50-95', color='green')
axes[0, 0].set_title('mAP50 and mAP50-95 over Epochs')
axes[0, 0].set_xlabel('Epoch')
axes[0, 0].set_ylabel('mAP')
axes[0, 0].legend()
axes[0, 1].plot(df['epoch'], df['metrics/precision(B)'], label='Precision', color='purple')
axes[0, 1].set_title('Precision over Epochs')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Precision')
axes[0, 1].legend()
axes[1, 0].plot(df['epoch'], df['metrics/recall(B)'], label='Recall', color='orange')
axes[1, 0].set_title('Recall over Epochs')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('Recall')
axes[1, 0].legend()
axes[1, 1].plot(df['epoch'], df['train/box_loss'], label='Train Box Loss', color='brown')
axes[1, 1].plot(df['epoch'], df['train/cls_loss'], label='Train Class Loss', color='red')
axes[1, 1].set_title('Training Loss over Epochs')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('Loss')
axes[1, 1].legend()


axes[2, 0].plot(df['epoch'], df['val/box_loss'], label='Validation Box Loss', color='magenta')
axes[2, 0].plot(df['epoch'], df['val/cls_loss'], label='Validation Class Loss', color='cyan')
axes[2, 0].set_title('Validation Loss over Epochs')
axes[2, 0].set_xlabel('Epoch')
axes[2, 0].set_ylabel('Loss')
axes[2, 0].legend()


plt.tight_layout(rect=[0, 0, 1, 0.96])


plt.show()
```
